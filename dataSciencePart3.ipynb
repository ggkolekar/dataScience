# 6. **Merging dataframes**

```python
df2 = pd.DataFrame([{'Name' : 'Setosa', 'Species' : 'Species 1'},
                  {'Name':'Versicolor','Species':'Species 2'},
                   {'Name':'Virginica','Species':'Species 3'}])
df2
```

# 8. **Aggregate function**

```python
df1.head()
```

```python
#Aggregate average of Sepal Length for each Species
df1.groupby('variety')['sepal.length'].mean()
```

```python
#Aggregate average of Sepal Length for each Species
df1.groupby('variety')['sepal.length','sepal.width','petal.length','petal.width'].mean()
```

```python
#Let's average the SepalLength and sum up the SepalWidth
a = df1.groupby('variety').agg({'sepal.length':'mean','sepal.width':'sum'})
a
```

```python
#Aggregate average of all the numerical columns for each Species
df1.groupby('variety')[['sepal.length','sepal.width','petal.length','petal.width']].mean()
```

```python
#Let's average the SepalLength and sum up the SepalWidth
a = df1.groupby('variety').agg({'sepal.length':'mean','sepal.width':'sum'})
a
```

```python
#Let's rename columns for sanity
a.rename(columns={'sepal.length':'Avg_sepal.length','sepal.width':'Sum_sepal.width'})
```

# 9. **Subsetting dataframes**

```python
#Subset data in pandas based on condition

sd=iris.rename(columns=({'variety':'Name','sepal.length':'SepalLength','sepal.width':'SepalWidth','petal.length':'PetalLength','petal.width':'PetalWidth'}))
sd.head()
```

```python
#Subset data in pandas based on two conditions
b = sd[(sd.SepalLength>5) & (sd.SepalLength>6)]
b.head()
```

```python
#Subset only for one Species
c = sd[sd['Name']=='Versicolor']
c.head()
```

```python
#Subset for two species
d=sd[sd['Name'].isin(['Virginica','Versicolor'])]
d.head()
```
