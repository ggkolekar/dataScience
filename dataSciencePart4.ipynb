# 2. **Conditional column**

```python
#Create new column for bucketing the Sepal Length as High or Low
a['SepalLength_Size'] = np.where(a['sepal.length']>=5,'High','Low')
a.head()
```

```python
#Using list comprehension
a['SepalLength_Size'] = ['High' if x >= 5 else 'Low' for x in a['sepal.length'] ]
a.head()
```

```python
#Using assign function
def size(row_number):
    if row_number["sepal.length"] >=5:
        return 'High'
    else:
        return 'Low'

a = a.assign(SepalLength_Size=a.apply(size, axis=1))

a.head()
```

```python
#Using lambda function
a['SepalLength_Size'] = a['sepal.length'].map( lambda x: 'High' if x >= 5 else 'Low')
a.head()
```

# 4. **Cross tab function**

```python
pd.crosstab(a['variety'],a['SepalLength_Size'])
```
